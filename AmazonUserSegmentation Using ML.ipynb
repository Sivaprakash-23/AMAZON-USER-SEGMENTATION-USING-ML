```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_set = pd.read_csv('/content/Amazonuser.csv')
X = data_set.iloc[:,[3,4]].values
```

```python
print(X)
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init="k-means++",random_state=21)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss)
plt.title('wcss via elbow method')
plt.xlabel('cluster')
plt.ylabel('wcss value')
plt.show()
```

```python
kmeans=KMeans(n_clusters=4,init="k-means++",random_state=20)
y_means=kmeans.fit_predict(X)

```

```python
print(y_means)
```

```python
plt.scatter(X[y_means== 0,0],X[y_means==0,1],s=100,c="magenta",label='cluster 1')
plt.scatter(X[y_means== 1,0],X[y_means==1,1],s=100,c="blue",label='cluster 2')
plt.scatter(X[y_means== 2,0],X[y_means==2,1],s=100,c="red",label='cluster 3')
plt.scatter(X[y_means== 3,0],X[y_means==3,1],s=100,c="cyan",label='cluster 4')
plt.title("CLUSTER OF AMAZON USER SEGMENT")
plt.xlabel("INCOME")
plt.ylabel("PURCHASE")
plt.legend()
plt.show()
```
